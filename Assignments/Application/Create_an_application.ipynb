In this assignment you have to create an application using one of the popular tools such as:
* Streamlit: https://streamlit.io/
* Chainlit: https://docs.chainlit.io/get-started/overview

Which uses a REST API built with:
* FastAPI: https://fastapi.tiangolo.com/
* Django: https://www.djangoproject.com/
* Flask: https://flask.palletsprojects.com/en/stable/
* JupyterKernelGateway: https://jupyter-kernel-gateway.readthedocs.io/en/latest/


Choose any of the previous assignments to create an application for its topic!

The application itself doesn't have to be too complicated, you just have to show that you can create one with 3-4 interactive features that either gives further possibilities to previously discussed topics or is about something completely different topic.
